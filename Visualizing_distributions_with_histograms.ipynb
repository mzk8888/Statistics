```python
import numpy as np #Numpyは数値計算を効率的に行うためのライブラリ
```

```python
import matplotlib.pyplot as plt #Matplptlibはグラフ描画用のライブラリ
```

```python
from scipy.stats import norm #SciPy.statsは統計解析用のライブラリ
```

```python
#パラメーターの設定
mu = 50 #平均50
sigma = 10 #標準偏差10
```

```python
#ランダムデータの生成（正規分布）
data = norm.rvs(loc=mu, scale=sigma, size=1000)
#norm.rvs()関数を用いて、平均50、標準偏差10の正規分布に従うランダムデータを1万個生成。この1万個のデータをヒストグラムで表示する
```

```python
#ヒストグラムの描画
plt.hist(data, bins=30, density=True, alpha=0.5, color='red')
#plt.hist()関数を用いて、この1万個のデータを30の区間に分けてヒストグラムを描画する
```

```python
#正規分布の確率密度関数から理論的な曲線の描画
xmin,xmax = plt.xlim()
x = np.linspace(xmin,xmax,100) #グラフにおけるX軸を設定
p = norm.pdf(x,mu,sigma)
plt.plot(x,p,'k',linewidth=2) #norm.pdf()関数を用いて理論的な曲線を計算し、plt.plot()関数を用いて描画
```

```python
title = f"Normal Distribution (mean:{mu:.1f},standard deviation:{sigma:1f})"
plt.title(title)
plt.show() #plt.title()関数を用いてグラフのタイトルを設定し、plt.show()関数を用いてグラフを表示する
```
